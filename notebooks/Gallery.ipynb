# Tutorial Gallery

Each tutorial is marked with its last-tested date to make it clear what state the notebooks are in. Testing is done at [NCI](https://nci.org.au/) with a data archive already established. Some notebooks also draw data from cloud hosted data sources.


## Low Hardware Requirements

These tutorials can be run on a 4GB GPU using relatively low volumes of data (3-10GB). They will also work in HPC environments.

- [Train and run a simplified global weather model (low hardware and data requirements)](./tutorial/FourCastMini_Demo.ipynb)  (significantly updated on 1 June 2025)
