# Tutorial Gallery

This gallery is organised into:

- **[Low Hardware Requirements and Quick Start](#Low-Hardware-Requirements-and-Quick-Start)**.  
  *Good for newcomers. Includes tips and tricks, and dealing with common questions when first adopting PyEarthTools.*  
- **[Tutorials on Specific Modelling Objectives](#Tutorials-on-Specific-Modelling-Objectives)**.  
  *Good starting points for people interested in specific modelling opportunities, or for those looking to see an end-to-end approach.*  
- **[How-to Guides for Working with Components of PyEarthTools](How-to-Guides-for-Components-of-PyEarthTools)**.  
  *Good for those wanting to understand a specific concept within PyEarthTools but don't require an end-to-end modelling demonstration.*
  - Introductory Guides
  - Deep Dive - The Data Module
  - Deep Dive - The Pipeline Module

Each tutorial is marked with its last-tested date. 

Manual testing is done at [NCI (Australia)](https://nci.org.au/) with a data archive already established. Some notebooks also draw data from cloud hosted data sources. Some notebooks are also tested in other computing environment when time allows.
